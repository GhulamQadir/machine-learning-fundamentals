# **Machine Learning Pipeline**

The general steps to solve a Machine Learning problem:

---

### 1. Data (Dataset)
- Collect or load dataset
- Split into Features (X) and Target (Y)

---

### 2. Exploratory Data Analysis (EDA)
- Check dataset shape (rows × columns)
- Identify missing values
- Detect outliers
- Look at feature distributions

---

### 3. Data Cleaning
- Fix incorrect datatypes
- Handle missing values
- Encode categorical variables (e.g., one-hot encoding)

---

### 4. Feature Engineering
- Create new features
- Scale/normalize numerical values
- Select important features

---

### 5. Splitting Data
- Split dataset into:
  - **Training set** (to train the model)
  - **Testing set** (to evaluate model performance)

---

### 6. Model Training
- Choose algorithm (Linear Regression, Decision Tree, Logistic Regression, etc.)
- Train the model on the training set

---

### 7. Evaluation
- Measure model performance using metrics:
  - Accuracy
  - Precision
  - Recall
  - Confusion Matrix
  - F1 Score
  - R² (for regression)
