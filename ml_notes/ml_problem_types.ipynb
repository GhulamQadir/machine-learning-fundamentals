# **Machine Learning Problem Types**

#### Can we predict someone has heart disease based on their health measurements?

---

### 1. Binary Classification
- Predict one of two possible categories
- Output: 0 or 1, Yes/No, True/False  

**Example:**  
- Predict whether a patient has heart disease (1 = Disease, 0 = No Disease)

---

### 2. Multi-Class Classification
- Predict one of more than two categories  

**Example:**  
- What type of animal is this? (Cat, Dog, Horse, Cow, Goat)

---

### 3. Regression
- Predict a continuous numerical value  

**Example:**  
- Predict the selling price of a house

---

### 4. Multi-Label Classification
- Predict multiple binary labels for each instance  

**Example:**  
- Predict which genres apply to a movie (Action, Comedy, Thriller, Drama)

---

### 5. Clustering (Unsupervised Learning)
- Group similar items together without labeled targets  

**Example:**  
- Group customers by spending behaviour

---

### 6. Anomaly Detection
- Find unusual data points that do not fit normal patterns  

**Example:**  
- Detecting credit card fraud

---

### 7. Time Series Forecasting
- Predict values over time  

**Example:**  
- Forecast stock prices / electricity usage over days, weeks, months

---

## Heart Dataset Example
- **Type:** Supervised Learning  
- **Problem:** Binary Classification  
- **Goal:** Predict Target (1: Disease, 0: No Disease)  
- **X (Input or Features:)** Input data (Age, Thal, Cholestrol, CPType, MaxHR....)<br>
- **Y (Output or Label:)** Output Column (Target)<br>
