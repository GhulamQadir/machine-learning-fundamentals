```python
"""
Regression: Predict continuous value
Classification: Predict category

Simple Linear regression: 
Predict y using one independent variable x
y = B0 + B1x + E
B0: intercept (value of y when x=0)
B1: Slope
E: Error term
"""
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

# **Single Linear Regression**
- You have one input (x) and one output (y).
- It tries to draw the best straight line through the data points.
- **Formula:** y=β0​+β1​x
- **Example:** predicting a student’s marks just from study hours.

```python
# Dataset
x = np.array([1,2,3,4,5]).reshape(-1,1)
y = np.array([35,37,45,50,60])

# Train model
model = LinearRegression()
model.fit(x,y)

# Predictions
y_pred = model.predict(x)

# Plot results
plt.scatter(x,y, color="blue", label="Actual Data")
plt.plot(x,y_pred, color="red", label="Best Fit Line")
plt.xlabel("Hours Studied")
plt.ylabel("Marks")
plt.legend()
plt.show()

print("Intercept B0: ", model.intercept_)
print("Slope B1: ", model.coef_)
```

# **Multiple Linear Regression**
- You have two or more inputs (x₁, x₂, x₃, …) and one output (y).
- It’s still a straight-line relationship, but in higher dimensions (a plane or hyperplane).
- **Formula:** y=β0​+β1​x1​+β2​x2​+⋯+βn​xn​
- **Example:** predicting a student’s marks from study hours, sleep hours, and attendance together.

```python
import pandas as pd

# 1. Create dataset
apartments_data = {
    "size_sqft": [1000,1500,2000,2500,3000],
    "bedrooms": [2,3,3,4,4],
    "location_score": [5,6,7,8,9],
    "price": [200000, 250000, 300000, 350000, 400000]
}

# Convert dictionary into a Pandas DataFrame
df = pd.DataFrame(apartments_data)

df.head()
```

```python
# 2. Define features (X) and target (y)
# -------------------------------
# Independent variables (features): size, bedrooms, location_score
x = df[["size_sqft", "bedrooms", "location_score"]]

# Dependent variable (target): price
y = df["price"]
```

```python
# Train model

# Create a Linear Regression model object
model  = LinearRegression()

# Fit (train) the model using the dataset (X and y)
model.fit(x,y)

# Print intercept (B0) of the regression equation
print("Intercept B0: ", model.intercept_)

# Print coefficients (slopes B1, B2, B3) for each feature
print("Slope B1, B2, B3: ", model.coef_)
```

```python
# Predict apartment prices using the trained model
y_pred = model.predict(x)

# Print predicted values
print(y_pred)
```

```python
# 5. Visualize results
# -------------------------------
# Scatter plot: Actual prices (y) vs Predicted prices (y_pred)
plt.scatter(y,y_pred, color="blue", label="Predicted vs Actual")

# Plot a red dashed 45° line (ideal case: predictions = actual values)
plt.plot([y.min(), y.max()],[y.min(), y.max()], color="red", linestyle="--")

# Add axis labels and title
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Multiple Linear Regression - Predicted vs Actual")

# Display legend box using the labels defined in the plots (e.g., "Predicted vs Actual")
plt.legend()
plt.show()
```
