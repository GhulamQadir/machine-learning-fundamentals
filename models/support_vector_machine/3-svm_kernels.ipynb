```python
from matplotlib import pyplot as plt
from sklearn.datasets import make_circles
from sklearn.datasets import make_moons
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

### 1. GENERATE CIRCULAR NON-LINEAR DATA

```python
# make_circles  creates circular-shaped non-linear dataset
# n_samples = total points
# factor = inner vs outer circle size difference
# noise = randomness added to make data realistic

X, y = make_circles(n_samples=200, random_state=0, factor=0.5, noise=0.1) 
plt.scatter(X[:,0], X[:,1], c=y, cmap="coolwarm")
plt.title("Non-Linear Data - Circular Pattern")
plt.show()
```

```python
# 2) TRAIN SVM WITH RBF KERNEL

# kernel="rbf"  Most powerful kernel for non-linear data
# C = Margin strictness (bigger C = fewer errors allowed)
model = SVC(kernel="rbf", C=1)
model.fit(X, y)
# RBF kernel automatically maps data to higher dimensions
# where circular data becomes linearly separable.
```

```python
# Predict on training data
y_pred = model.predict(X)

# Calculate accuracy
print("RBF Kernel Accuracy", accuracy_score(y, y_pred))
```

```python
# Plot predictions (SVM output)
# c=y_pred  colors represent predicted classes
plt.scatter(X[:, 0], X[:,1], c=y_pred, cmap="bwr")
plt.title("SVM with RBF Kernel")
plt.show()
```

### 3. MAKE MOONS DATA (ANOTHER NON-LINEAR PATTERN)

```python
X, y = make_moons(n_samples=200, noise=0.1, random_state=0)

# SVM with RBF kernel
# kernel="rbf"  good for curved / non-linear decision boundaries
# C=1  normal strictness (not too strict, not too relaxed)
# gamma=0.5  controls how much influence each point has:
# low gamma = smoother boundary, high gamma = tighter curves
model = SVC(kernel="rbf", C=1, gamma=0.5)
model.fit(X, y)
y_pred = model.predict(X)
print("Accuracy: ", accuracy_score(y, y_pred))
```

```python
# Points colored by their predicted classes
plt.scatter(X[:, 0], X[:,1], c=y_pred, cmap="bwr")
plt.title("SVM with RBF Kernel")
plt.show()
```

```python
# Creating circular (donut-shaped) non-linear data
X, y = make_circles(n_samples=200, random_state=0, factor=0.5, noise=0.05)

# Polynomial Kernel SVM
# kernel="poly"  polynomial transformation
# degree=3  curve complexity (3rd degree polynomial)
# C=1  normal strictness
model = SVC(kernel="poly", degree=3, C=1)
model.fit(X, y)
y_pred = model.predict(X)
print("Accuracy:", accuracy_score(y, y_pred))
plt.scatter(X[:,0], X[:,1], c=y_pred, cmap="coolwarm")
plt.title("SVM with Polynomial Kernel on Circular Data")
plt.show()
```
