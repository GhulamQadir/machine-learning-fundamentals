## **Support Vector Machine (SVM)**
- Support Vector Machine (SVM) is a supervised machine learning algorithm used for:<br>
1. **Classification** (main use)<br>
2. **Regression** (less common)<br>
- SVM works by finding a decision boundary (hyperplane) that best separates the classes in the feature space.

- The key components of SVM include:
1. **Support Vectors**: 
- Data points that are closest to the decision boundary (hyperplane).
- These points define the position of the hyperplane.
- Only these points affect the SVM; other points do not.

2. **Hyperplane**:
- A line (2D), plane (3D), or general n-dimensional plane that separates the classes.<br>

3. **Margin**<br>
- Distance between the hyperplane and the closest points from each class (support vectors).
- SVM‚Äôs goal is to maximize this margin for better generalization.

4. **Maximum Margin**<br>
- SVM selects the hyperplane with largest possible distance from the nearest data points of each class.
- A bigger margin ‚Üí more confident & robust classification.

5. **Linear vs Non-linear SVM**<br>
- Linear SVM ‚Üí data can be separated by a straight hyperplane.
- Non-linear SVM ‚Üí uses kernels (RBF, polynomial) to separate data in higher-dimensional space.

#### **Examples:**<br>
1. Email Spam Filtering
2. Face Recognition
3. Medical Diagnosis
4. Credit Card Fraud Detection

## **Main Points:**

- Each Feature (or column) = One Dimension
- A hyperplane is the boundary that separates data
    - in 2D, it's a line
    - in 3D, it's a flat plane
    - in higher dimensions, it's a hyperplane

- Nearest points to the separation line are called support vectors
- Support vectors are the closest points from each class to the dividing line
- They are the ones that define the position of the boundary.
- If we remove SV, the line shifts.
- If we remove others, the line barely changes
- If new data appears near the margin, model has low confidence in it. 
- If the new data appears far from the margin, the model is highly confident.

### **SVM can perform three types of tasks:**
    1. Classification (mostly used for classification)
    2. Regression
    3. Outlier Detection

- SVC(Support Vector Classifier) for Classification
- SVR(Support Vector Regression) for Regression

1. **SVC(kernel='linear')** ‚Üí Slow, accurate, small dataset
2. **LinearSVC** ‚Üí Fast, scalable, large dataset
